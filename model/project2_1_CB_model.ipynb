```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/train_data.csv')
df.movie_title = df.movie_title.str[:-1]
```

```python
df
```

```python
df_pre = df.iloc[:,2:-4]
df_pre = df_pre.drop(['timestamp', 'release_date'], axis=1)
df_pre
```

```python
counting = df_pre.groupby('movie_title')['movie_title'].count()
counting
```

```python
ratings = df_pre.groupby('movie_title')['rating'].mean()
ratings
```

```python
df_movie = df_pre.iloc[:, 1:]
df_movie = df_movie.groupby('movie_title').mean().astype('int')
df_movie['rating'] = ratings
df_movie['count'] = counting
df_movie
```

```python
# 리뷰가 50개 이상인 영화만 (상위 64%)
m = df_movie['count'].quantile(0.64)
m
```

```python
df_movie_50 = df_movie.loc[df_movie['count'] >= m]
df_movie_50
```

```python
# 전체영화평균평점
a = df_movie_50['rating'].mean()
a
```

```python
def weight_rating(x, m=m, a=a):
    v = x['count']
    R = x['rating']
    return (v / (v+m) * R) + (m / (m+v) * a)
```

```python
df_movie_50['score'] = df_movie_50.apply(weight_rating, axis=1)
df_movie_50
```

```python
df_genre = df_movie_50.iloc[:, :-3]
df_genre
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cb_c_sim = cosine_similarity(df_genre, df_genre)
cb_c_sim = pd.DataFrame(cb_c_sim, df_genre.index, df_genre.index)
cb_c_sim
```

```python
cb_c_sim['12 Angry Men'].sort_values(ascending=False)[:20].drop('12 Angry Men')
```

```python
def get_recommand_movies(movie_title, cos_x=cb_c_sim, top=20):
    # 입력 영화 삭제
    sim_df = cos_x[movie_title].sort_values(ascending=False).drop(movie_title)
    # top순위만큼 추천
    sim_df = sim_df[:top]
    
    return sim_df.index
    
```

```python
get_recommand_movies('12 Angry Men')
```

```python
test=pd.read_csv('data/test_data.csv')
test.movie_title=test.movie_title.str[:-1]
```

```python
def test_score():
    precision_list=[]
    recall_list=[]
    for i in test['user_id'].unique():
      test_title=list(test[(test.rating>3)&(test.user_id==i)].movie_title)
      # 무작위로 행 순서 바꾸기
      # test_title = np.random.shuffle(test_title)
      try:
        if len(test_title)>1:
          recommended_list=list(get_recommand_movies(test_title[0]))
          count=0
          for value in test_title[1:]:
            if value in recommended_list:
              count+=1
          # 추천목록에 입력영화가 있는 경우 개수 1 삭제
          if test_title[0] in recommended_list:
            r = len(recommended_list) - 1
          else:
            r = len(recommended_list)
            
          precision=count/r
          recall=count/len(test_title)
          precision_list.append(precision)
          recall_list.append(recall)

      except:
        pass

    return f'precision: {np.mean(precision_list)}, recall: {np.mean(recall_list)}'
```

```python
test_score()
```
