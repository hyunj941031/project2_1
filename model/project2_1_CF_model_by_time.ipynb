```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
data=pd.read_csv('/content/drive/MyDrive/project_2_1/data/train_data_by_time.csv')
```

```python
data
```

```python
data.movie_title=data.movie_title.str[:-1]
```

```python
#유저와 영화에 대한 평점 행렬 생성 
```

```python
rating_table=data.pivot_table('rating','movie_title','user_id')
```

```python
rating_table=rating_table.fillna(0)
```

```python
rating_table
```

```python
# 코사인 유사도 계산 
```

```python
similarity=cosine_similarity(rating_table)
```

```python
similarity
```

```python
#코사인 유사도 데이터 활용하여 영화간 유사성 추출 
```

```python
similarity_table=pd.DataFrame(similarity,rating_table.index,rating_table.index)
```

```python
similarity_table
```

```python
def run(title):
    return similarity_table[title].sort_values(ascending=False)[1:21]
```

```python
run('English Patient, The')
```

```python
#평가 진행 
```

```python
test=pd.read_csv('/content/drive/MyDrive/project_2_1/data/test_data_by_time.csv')
test.movie_title=test.movie_title.str[:-1]
```

```python
test
```

```python
#학습되지 않은 데이터에 대하여 예측된 다른 영화를 실제로 선호하는지 측정
```

```python
def test_score():
    precision_list=[]
    recall_list=[]
    for i in test['user_id'].unique():
      test_title=list(test[(test.rating>3)&(test.user_id==i)].sort_values('timestamp').movie_title)
      try:
        if len(test_title)>1:
          recommended_list=list(run(test_title[0]).index)
          count=0
          for value in test_title[1:]:
            if value in recommended_list:
              count+=1
          precision=count/len(recommended_list)
          recall=count/(len(test_title)-1)
          precision_list.append(precision)
          recall_list.append(recall)

      except:
        pass

    return f'precision: {np.mean(precision_list)}, recall: {np.mean(recall_list)}'
```

```python
test_score()
```
