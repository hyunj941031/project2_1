```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
data=pd.read_csv('/content/drive/MyDrive/project_2_1/data/train_data.csv')
```

```python
data
```

```python
data.movie_title=data.movie_title.str[:-1]
```

```python
rating_table=data.pivot_table('rating','movie_title','user_id')
```

```python
rating_table=rating_table.fillna(0)
```

```python
rating_table
```

```python
similarity=cosine_similarity(rating_table)
```

```python
similarity
```

```python
similarity_table=pd.DataFrame(similarity,rating_table.index,rating_table.index)
```

```python
similarity_table
```

```python
def run():
  title=input('영화 제목을 입력해주세요 ')
  if title in similarity_table.columns:
    return similarity_table[title].sort_values(ascending=False)[1:11]
  else:
    return '정확한 영화 제목을 입력해주세요'
```

```python
run()
```

```python
test=pd.read_csv('/content/drive/MyDrive/project_2_1/data/test_data.csv')
test.movie_title=test.movie_title.str[:-1]
```

```python
def test_score():
    score_list=[]
    for i in test['user_id'].unique():
      test_title=list(test[(test.rating>3)&(test.user_id==i)].movie_title)
      try:
        if len(test_title)>1:
          score=similarity_table[test_title[0]][similarity_table[test_title[0]].index.isin(test_title[1:])].mean()
          score_list.append(score)
      except:
        pass

    return np.mean(score_list)
```

```python
test_score()
```
