```python
import pandas as pd

data = pd.read_csv('data/full_data.csv')
data = data.sort_values(by='timestamp').reset_index(drop=True)
data
```

```python
len(data) *0.8
```

```python
# 79992번째와 79993번째 데이터의 timestamp가 같음
train_data = data.iloc[:79993]
test_data = data.iloc[79993:].reset_index(drop=True)
```

```python
train_data.to_csv('data/train_data_by_time.csv', sep=',', index=False)
test_data.to_csv('data/test_data_by_time.csv', sep=',', index=False)
```
